Los requisitos para que funcione de forma automática son los siguientes:
- Líneas de código: cada línea, paquetería y librería tendrá su explicación de uso en la automatización.
- Datos: Los datos fueron creados para este trabajo, son pocos datos pero esto es escalable a la cantidad de datos deseados o necesarios.
- Plantilla: Se redactó una plantilla de carta a proveedores, esta será llenada con las líneas de código estableciadas.

```python
import pandas as pd
from datetime import datetime
from docxtpl import DocxTemplate
```

```python
#Lectura de la plantilla realizada para ser automatizada.
doc = DocxTemplate('plantilla.docx')
```

```python
#Declaración de variables.
name = 'Diffel Villalva Anguiano'
phone = '44-33-57-32-48'
mail = 'villalva.diffel@gmail.com'
date = datetime.today().strftime('%d/%m/%Y')
```

```python
#Creación de diccionario con variables
constantes = {'name':name,'phone':phone,'mail':mail,'date':date}
```

```python
df = pd.read_excel('data.xlsx')
df
```

```python
#ciclo for que itera sobre cada una de las columnas para darles un valor.
for indice, col in df.iterrows():
    contenido ={
        'company':col['company'],
        'material':col['material'],
        'quantity':col['quantity']
    }
    contenido.update(constantes)

    #Hace que el contenido se meta a cada variable en la plantilla
    doc.render(contenido)
    #Lo guarda con el nombre especificado
    doc.save(f'Proveedor {col['company']}.docx')
```
